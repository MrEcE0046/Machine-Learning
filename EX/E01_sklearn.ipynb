````python
import pandas as pd
import numpy as np
import seaborn as sns

df = sns.load_dataset("mpg")
df = df.reset_index()
df.head(10)
````

````python
df.describe()
````

````python
df.info()
````

````python
datatype = {"quant": ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year"], 
            "qual": ["origin", "name"]}

quants = df[datatype["quant"]].astype(np.float_)
df = pd.concat([quants, df[datatype["qual"]]], axis=1)
df
````

````python
df.dropna(inplace=True)
X = df.drop(columns=["origin","name"])
# df.corr()["mpg"]
````

````python
X, y = X.drop("mpg", axis=1), df["mpg"]
X.insert(0, "Intercept", 1)
y.shape
````

## 2. Function for evaluation (*)

Create a function for training a regression model, predicting and computing the metrics MAE, MSE, RMSE. It should take in parameters of X_train, X_test, y_train, y_test, model. Now create a linear regression model using scikit-learns ```LinearRegression()``` (OLS normal equation with SVD) and call your function to get metrics. 

<details>

<summary>Answer</summary>

MAE 2.50

MSE 10.50

RMSE 3.24

</details>

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error

def training_model(X_train, X_test, y_train, y_test):
    model = LinearRegression()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    
    MAE = mean_absolute_error(y_test, y_pred)
    MSE = mean_squared_error(y_test, y_pred)
    RMSE = root_mean_squared_error(y_test, y_pred)
    
    return MAE, MSE, RMSE

lr = training_model(X_train, X_test, y_train, y_test)
print(lr)
````

````python
from sklearn.preprocessing import StandardScaler

""" Genom att skala data innan modellen tränas, kan du förbättra modellens prestanda 
och minska risken för överfitting. """

def scaled_training_model(X_train, X_test, y_train, y_test):
    scaler = StandardScaler()
    scaled_X_train = scaler.fit_transform(X_train)
    scaled_X_test = scaler.transform(X_test)

    model = LinearRegression()
    model.fit(scaled_X_train, y_train)

    y_pred = model.predict(scaled_X_test)
    
    scaled_MAE = mean_absolute_error(y_test, y_pred)
    scaled_MSE = mean_squared_error(y_test, y_pred)
    scaled_RMSE = root_mean_squared_error(y_test, y_pred)
    
    return scaled_MAE, scaled_MSE, scaled_RMSE

scaled_lr = scaled_training_model(X_train, X_test, y_train, y_test)

df_results = pd.DataFrame(scaled_lr)
df_results.insert(0, "", ["MAE", "MSE", "RMSE"])
df_results.rename(columns={df_results.columns[1]: "LR scaled"}, inplace=True)
df_results.insert(1, "LR", [lr[0], lr[1], lr[2]])  
````

````python
from sklearn.preprocessing import PolynomialFeatures

def polynomial_regression(X_train, X_test, y_train, y_test):
    results = []
    deg = (1, 2, 3)
    for i in deg:
        poly_features = PolynomialFeatures(degree=i)
        poly_X_train = poly_features.fit_transform(X_train)
        poly_X_test = poly_features.transform(X_test)

        model = LinearRegression()
        model.fit(poly_X_train, y_train)

        y_pred = model.predict(poly_X_test)
        
        poly_MAE = mean_absolute_error(y_test, y_pred)
        poly_MSE = mean_squared_error(y_test, y_pred)
        poly_RMSE = root_mean_squared_error(y_test, y_pred)
        
        results.extend([[poly_MAE, poly_MSE, poly_RMSE]])  

    return results

poly = polynomial_regression(X_train, X_test, y_train, y_test)

for i, result in enumerate(poly):
    df_results.insert(3, f"LR Poly {3 - i}", poly[2-i])

df_results
````
